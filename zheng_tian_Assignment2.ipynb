### 1 Iris Dataset [35 points]

```python
import pandas as pd
fileURL = "http://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
df = pd.read_csv(fileURL, names=["Sepal Length", "Sepal Width", "Petal Length", "Petal Width", "Name"], header=None)
```

__1.1 Summary Statistics__

Print the first 5 elements of your DataFrame using the command head(). How many features are there and what are their types (e.g., numeric, nominal)?

```python
df.head()
```

There are 5 features: sepal length, sepal width, petal length, petal width, and name. Sepal length, sepal width, petal length, and petal width are numeric types; they are numeric values in unit of cm. Name is a nominal type; name is one class out of Iris Setosa, Iris Versicolour, or Iris Virginica.

Compute and display summary statistics for each feature available in the dataset. These must include the minimum value, maximum value, mean, range, standard deviation, variance, count, and 25:50:75% percentiles.

```python
summary = df.describe(include='all')
summary
```

```python
# Display the minimum value, maximum value, mean, range, standard deviation, variance, count, and 25:50:75% percentiles
# Extract data from summary
# Range is calculated by subtracting minimum value from maximum value
pd.DataFrame([summary.iloc[6], summary.iloc[10], summary.iloc[4], summary.iloc[10] - summary.iloc[6], 
              summary.iloc[5], df.var(), summary.iloc[0], summary.iloc[7], summary.iloc[8], summary.iloc[9]], 
             index=['Min', 'Max', 'Mean', 'Range', 'Standard deviation', 'Variance', 'Count', '25%', '50%', '75%'])
```
